```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
c1 = torch.load('./tmp/c1.pt')
c2 = torch.load('./tmp/c2.pt')
loss_arr = torch.load('./tmp/loss_arr.pt')
```

```python
iter_arr = np.array(loss_arr['Iter'])
D_loss_arr = np.array(loss_arr['D_loss'])
G_loss_arr = np.array(loss_arr['G_loss'])
```

```python
Dloss, = plt.plot(iter_arr, D_loss_arr, label='D_loss')
Gloss, = plt.plot(iter_arr, G_loss_arr, label='G_loss')
plt.legend()
plt.show()
```

```python
import librosa
```

```python
y, sr = librosa.load('data/nottingham-dataset/wav/ashover1.wav')
```

```python
y_third_up = librosa.effects.pitch_shift(y, sr, n_steps=4)
```

```python
y_gen, _ = librosa.load('tmp/gen.wav')
```

```python
plt.plot(y, alpha=0.4, label='Raw Music')
plt.plot(y_third_up, alpha=0.4, label='Shifted 3-key Higher')
plt.plot(y_gen, alpha=0.4, label='Generateed Music')
plt.legend()
plt.show()
```
